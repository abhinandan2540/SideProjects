```python
# solving logistic growth model
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
r=1
K=1000
X0=15
t=np.linspace(0,50,100)

def logistic_growth(X0,t):
    dXdt=r*X0*(1-(X0/K))
    return dXdt

X=odeint(func=logistic_growth, y0=X0, t=t)
plt.plot(t, X, label='logistic growth')
plt.xlabel('T')
plt.ylabel('X(t)')
plt.title('dX/dt=rX(1-X/K)')
plt.grid()
plt.show()
```

```python
# limited growth in harvesting
r=1 # reproduction rate
K=1000 # carrying capacity
h=0.01 # harvesting rate
X0=15 # initial population
t=np.linspace(0,100,100)

def limited_growth_harvesting(X0, t):
    dXdt = r*X0*(1-(X0/K)) - h
    return dXdt

X=odeint(func=limited_growth_harvesting, y0=X0, t=t)
plt.plot(t, X)
plt.show()
```

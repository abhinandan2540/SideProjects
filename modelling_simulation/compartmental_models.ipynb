```python
# solving differential equation with python
# exponential decay model
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt

# initial given conditions
K=2.0
N0=10^5
t=np.linspace(0,5,100) # generating 100 points from 0 to 5

def decay_model(N,K):
    dNdt=-K*N
    return dNdt

y=odeint(func=decay_model, y0=N0,t=t)
plt.plot(t,y, label='exponential decay')
plt.legend()
plt.xlabel('T')
plt.ylabel('N(t)')
plt.title('dN/dt=-KN')
plt.grid(True)
plt.show()
```

```python
# lake pollution model
F=4*12
Cin=3
V=28
C0=10
t=np.linspace(0,1,100)

def lack_pollution(C0,t):
    dCdt=(Cin-C0)*(F/V)
    return dCdt

y=odeint(func=lack_pollution,y0=C0,t=t)
plt.plot(t,y,label='lack pollution')
plt.legend()
plt.xlabel('T')
plt.ylabel('C')
plt.title('dC/dt=(Cin-C)*F/V')
plt.grid()
```

```python
# how long does it take to reach 5% of it's initial value
import math
F=480*10**6
V=458*10**9
def taking_time(C):
    t=-math.log(C)*(V/F)
    return t/365

print(taking_time(C=0.05))
```

```python
# drug assimilation model
K1=1.3860
K2=0.1386
X0=1
Y0=0
U0=[X0,Y0]
t=np.linspace(0,15,100)
def drug_assimilation_decongestant(U0,t):
    X0=U0[0]
    Y0=U0[1]
    dXdt=-K1*X0
    dYdt=K1*X0-K2*Y0
    return dXdt,dYdt

solution=odeint(func=drug_assimilation_decongestant,y0=U0,t=t)
X=solution[:,0]
Y=solution[:,1]
plt.plot(t, X, label='X (Drug in GI tract)')
plt.plot(t, Y, label='Y (Drug in bloodstream)')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Amount')
plt.title('Drug Assimilation Model')
plt.show()
```

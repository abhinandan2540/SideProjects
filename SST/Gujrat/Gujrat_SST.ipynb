```python
import pandas as pd
import numpy as np

df = pd.read_csv("2011.txt")
df.columns = ["DATE","TIME","LON","LAT","HT","SST"] # colums for csv
df["DATE"]= pd.to_datetime(df["DATE"], format="%d-%b-%Y") # DATE formatting

# Converting SST into Numeric values
df["SST"]=pd.to_numeric(df["SST"], errors="coerce")
# Replacing bad values with NaN
df["SST"] = df["SST"].replace(-1e34, np.nan)

df = df.dropna(subset=["SST"])

df_first=df[df["DATE"].dt.day ==1] # Keeping the first date of each month
df_first["MONTH"] = df_first["DATE"].dt.to_period("M") # Creation of month column

# Mean monthly SST
monthly_mean_SST = df_first.groupby('MONTH')["SST"].mean().reset_index()
monthly_mean_SST.to_csv("2011_Monthly_Mean_SST.csv", index=False)
```
